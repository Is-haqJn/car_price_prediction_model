Encoding the data

```python
x = df.iloc[:, [1,3,4,6]].values
y = df.iloc[:, 2].values
```

```python
x
```

```python
x[:,2]
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
x[:, 2] = lb.fit_transform(x[:, 2])
lb1 = LabelEncoder()
x[:, 3] = lb1.fit_transform(x[:, 3])
```

```python
x[:, 2]
```

SPLITTING THE DATA INTO TRAINING AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05, random_state=0)

print(x_train)
```

TRAINING THE MODEL

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=300, random_state=0)
regressor.fit(x_train, y_train)
```

```python
accuracy = regressor.score(x_test, y_test)
print(accuracy*100, '%')
```

Performing a Test on given input

```python
new_data = [ 2017, 70000, 'Petrol', 'Manual']
new_data[2] = lb.transform([new_data[2]])[0]
new_data[3] = lb1.transform([new_data[3]])[0]
```

```python
print(new_data)
regressor.predict([new_data])
```

Saving the model and encoder using Pickle

```python
import pickle
pickle.dump(regressor, open('car_price_regressor.pkl', 'wb'))
pickle.dump(lb, open('lb.pkl', 'wb'))
pickle.dump(lb1, open('lb1.pkl', 'wb'))
```
